```python
from bs4 import BeautifulSoup  ###解析HTML
from urllib.request import urlopen  ###下载网站内容，也就是html格式的文本
import re   ###正则表达式
from dateutil import parser  ###没用
import urllib  ###下载网站内容
```

```python
link = 'https://bitcoinist.com/page/2/?s&lang=en' ###网站上的搜索页
###把网站下载下来
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7' ###伪装成浏览器下载网站内容
headers={'User-Agent':user_agent} 
request = urllib.request.Request(link, None, headers) #The assembled request
response = urllib.request.urlopen(request)
html = response.read() # The data u need
###解析html
soup = BeautifulSoup(html, 'html.parser')
```

```python
###搜寻页面上可得到的链接
i = 1
for h3 in soup.findAll('h3', {'class': 'title'}):
    try:
        print(h3.a.get('href'))
        print(i)
        i += 1
    except:
        continue
```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'

link = 'https://bitcoinist.com/jpmorgan-bitcoin-dystopia/'
headers={'User-Agent':user_agent} 

request = urllib.request.Request(link, None, headers) #The assembled request
response = urllib.request.urlopen(request)
html = response.read() # The data u need
soup = BeautifulSoup(html, 'html.parser')
```

```python
### 爬取文章的时间 
time = soup.findAll('p')[2]
re.sub('<.*?>', '', str(time))[-22:]
```

```python
### 文章标题
soup.title
```

```python
print(soup.prettify())
```

```python
### 爬取文章内容
article = ''
for line in soup.findAll('p')[3:-23]:
    subline = re.sub('<.*?>', '', str(line))
    if len(subline) < 5:
        continue
    article += subline
    article += '\n'
print(article)
```
