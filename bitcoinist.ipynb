```python
from bs4 import BeautifulSoup
import re
import urllib
import pickle
import pandas as pd
```

```python
### 保存文章数据及内容
def init(file_name):
    try:
        # Restore articles
        with open(file_name, 'rb') as file:
            data = pickle.load(file)
            return data
    except FileNotFoundError:
        # No article downloaded
        data = {
            'current_page': 11, ###从该网站汇总页第11页开始都是2018年
            # article = {url: (title, time, content)}
            'articles': dict()  ###用url作为dictionary的key以保证爬取链接不重复
        }
        file = open(file_name, 'wb')
        pickle.dump(data, file)
        file.close()
        return data
```

```python
### 针对包含文章的网页，爬取时间、标题及内容。
def parse_article(url):
    print(url)
    
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers = {'User-Agent': user_agent} 
    request = urllib.request.Request(url, None, headers)
    response = urllib.request.urlopen(request)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    
    title = str(soup.title.string)
    
    time = soup.findAll('p')[2]
    time = re.sub('<.*?>', '', str(time))[-22:]
    
    content = ''
    for line in soup.findAll('p')[3:-23]:
        paragraph = re.sub('<.*?>', '', str(line))
        if len(paragraph) < 5:
            continue
        content += paragraph
        content += '\n'
        
    return title, time, content
```

```python
### 爬取文章汇总页的链接
def get_new_urls(page):
    urls = list()
    stream_url = 'https://bitcoinist.com/page/' + str(page) + '/?s&lang=en'
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers = {'User-Agent': user_agent} 
    request = urllib.request.Request(stream_url, None, headers)
    response = urllib.request.urlopen(request)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    
    for h3 in soup.findAll('h3', {'class': 'title'}):
        try:
            url = h3.a.get('href')
            urls.append(url)
        except:
            continue
    return urls
```

```python
file_name = 'bitcoinist.data'
```

```python
data = init(file_name)
end_page = 211  ###手动查找最后包含2018文章的页数
start_page = data['current_page'] + 1

for page in range(start_page, end_page+1):
    print("Start scraping page %d"%(page))
    urls = get_new_urls(page)
    for url in urls:
        # Check whether the article had been downloaded
        if url not in data['articles']:
            title, time, content = parse_article(url)
            data['articles'][url] = (title, time, content)
    # Save data
    data['current_page'] = page
    file = open(file_name, 'wb')
    pickle.dump(data, file)
    file.close()
    print("End scraping page %d"%(page))
    print()
```
